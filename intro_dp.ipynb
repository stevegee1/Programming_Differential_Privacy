```python
# Let's assume we are dealing with a counting query
# of rows with a particular property, in a dataset:
# 1. Max sensitivity = 1 (i.e, when we modify exactly one row,
#    the query's output can change by at most 1
# 2. We pick epsilon = 0.1
```

```python
import pandas as pd
import numpy as np
```

```python
# initialisation
adult = pd.read_csv("adult_with_pii.csv")
sensitivity = 1
epsilon = 0.1
```

```python
# Add noise
adult[adult['Age'] >= 40 ].shape[0] + np.random.laplace(loc=0, scale = sensitivity / epsilon)
```

```python

Karries_row = adult[adult['Name'] == 'Karrie Trusslove'].shape[0]
```

```python
Karries_row # This violates the privacy of karrie's income
```

```python
# To ensure differential privacy for the counting query
Karries_row_with_DP = adult[adult['Name'] == 'Karrie Trusslove'].shape[0] + np.random.laplace(loc=0, scale = sensitivity/epsilon)
```
